```python
!pip install pandas
!pip install matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
plot_df = pd.read_csv("task1_Q3.csv")
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(
    plot_df['longitude_rounded'],
    plot_df['latitude_rounded'],
    s=plot_df['count'] / 50,       
    alpha=0.6,
    color='#4A90E2',               
    edgecolors='white', linewidth=0.5)
# Styling
plt.title("Geographical Distribution of Tweets ", fontsize=14)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
# Save & Show
plt.savefig("task1_Q3.png")
plt.show()
```
