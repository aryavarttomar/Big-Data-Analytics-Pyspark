```python
!pip install pandas
!pip install matplotlib
import pandas as pd
import matplotlib.pyplot as plt
# Load the CSV file
df = pd.read_csv("task1_Q4.csv")  # Adjust path if needed
order = ["Morning", "Afternoon", "Evening", "Night"]
df_sorted = df.set_index("time_period").loc[order].reset_index()
# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(df_sorted['time_period'], df_sorted['count'], color='orange')
plt.title('Number of Tweets by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Tweets')
plt.grid(axis='y', linestyle='--', alpha=0.7)
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:,}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.savefig("task1_Q4.png", dpi=300)
plt.show()
```
