```python
!pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load genre rating data
df = pd.read_csv("task2_Q5.csv")  
# Sort genres by count 
df = df.sort_values("rating_count", ascending=False)
# Set up the plot
plt.figure(figsize=(12, 6))
bars = plt.bar(df["genre"], df["rating_count"], color="skyblue")
# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:,}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=9)
# Labels and title
plt.xlabel("Genre")
plt.ylabel("Number of Ratings")
plt.title("Number of Ratings per Genre")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
# Save plot to file
plt.savefig("task2_Q5.png", dpi=300)
print("task2_Q5.png'")
# Display plot
plt.show()
```
