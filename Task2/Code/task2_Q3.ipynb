```python
!pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("task2_Q3.csv") 
# Plot
plt.figure(figsize=(6, 6))
plt.pie(df["rating_count"], labels=df["rating_group"], autopct="%1.1f%%", startangle=140)
plt.title("Rating Distribution")
plt.axis("equal")
plt.tight_layout()
plt.savefig('task2_Q3.png', dpi=300, bbox_inches='tight')
print("Saved: task2_Q3.png")
plt.show()
```
