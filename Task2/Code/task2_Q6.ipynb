```python
!pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load your downloaded CSV
df = pd.read_csv("task2_Q6.csv") 
# Sort by year 
df = df.sort_values("year")
# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(df["year"], df["rating_count"], marker='o', linewidth=2)
# Annotate data points
for x, y in zip(df["year"], df["rating_count"]):
    plt.text(x, y + 2000, f'{y:,}', ha='center', fontsize=9)
# Titles and labels
plt.title("Number of Ratings per Year", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Ratings")
plt.grid(True)
plt.tight_layout()
# Save the plot
plt.savefig("task2_Q6.png", dpi=300)
print("task2_Q6.png'")
# Show plot
plt.show()
```
