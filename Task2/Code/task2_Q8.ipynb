```python
!pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("task2_Q8.csv")  
# Sort 
df = df.sort_values("user_category")
# Plot setup
plt.figure(figsize=(6, 4))
colors = ['#4daf4a', '#e41a1c']  
bars = plt.bar(df["user_category"], df["user_count"], color=colors)
# Add exact counts on top of bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:,}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=10)
# Labels and title
plt.title("User Activity Distribution", fontsize=14)
plt.xlabel("User Category")
plt.ylabel("Number of Users")
plt.tight_layout()
# Save the chart
plt.savefig("task2_Q8.png", dpi=300)
print("task2_Q8.png'")
# Show plot
plt.show()
```
