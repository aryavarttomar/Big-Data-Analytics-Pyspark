```python
!pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("task2_Q4.csv") 
# Set up the bar chart
fig, ax = plt.subplots(figsize=(7, 5))
# Bar colors
colors = ['#1f77b4', '#ff7f0e']  # Blue, Orange
bars = ax.bar(df["time_of_year"], df["rating_count"], color=colors)
# Add labels above bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:,}',  
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 5),
                textcoords="offset points",
                ha='center', va='bottom', fontsize=10)
# Add title and labels
ax.set_title("Ratings Distribution by Time of Year", fontsize=14)
ax.set_xlabel("Time of Year")
ax.set_ylabel("Number of Ratings")
plt.tight_layout()
# Save image
plt.savefig("task2_Q4.png", dpi=300, bbox_inches='tight')  
print("task2_Q4.png'")
# Show the plot
plt.show()
```
