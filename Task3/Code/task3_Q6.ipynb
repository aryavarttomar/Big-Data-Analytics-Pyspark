```python
!pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Load CSVs
unweighted_df = pd.read_csv("task3_Q6_unweighted.csv")
weighted_df = pd.read_csv("task3_Q6_weighted.csv")
# Unweighted PageRank Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(unweighted_df['id'], unweighted_df['pagerank'], color='skyblue')
plt.title("Top 5 Community Areas (Unweighted PageRank)")
plt.xlabel("Community Area ID")
plt.ylabel("PageRank Score")
plt.tight_layout()
plt.savefig("task3_Q6_unweighted.png")
plt.show()
# Weighted PageRank Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(weighted_df['id'], weighted_df['pagerank'], color='orange')
plt.title("Top 5 Community Areas (Weighted PageRank - Fare)")
plt.xlabel("Community Area ID")
plt.ylabel("PageRank Score")
plt.tight_layout()
plt.savefig("task3_Q6_weighted.png")
plt.show()
```
